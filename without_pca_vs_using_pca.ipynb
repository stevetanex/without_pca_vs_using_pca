```python
# first without pca using 785 data columns
x=df.iloc[:,1:]
y=df.iloc[:,0]
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=0,test_size=0.3)
X_train.shape,X_test.shape,Y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
y_predit=knn.predict(X_test)

accuracy_score(y_predit,Y_test)*100
```

```python

# # using PCA
# first std
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# import pca
from sklearn.decomposition import PCA
pc=PCA(n_components=100)# random nunber kudukka
X_train_pca=pc.fit_transform(X_train)
X_test_pca=pc.transform(X_test)

X_train_pca.shape,X_train.shape
```

```python
knn1=KNeighborsClassifier(n_neighbors=5)
knn1.fit(X_train_pca,Y_train)
y_predit_pca=knn1.predict(X_test_pca)

accuracy_score(y_predit_pca,Y_test)*100
```

```python
for i in range(1,785):
    pc=PCA(n_components=i)
    X_train_pca=pc.fit_transform(X_train)
    X_test_pca=pc.transform(X_test)
    knn1.fit(X_train_pca,Y_train)
    y_predit_pca=knn1.predict(X_test_pca)

    print(accuracy_score(y_predit_pca,Y_test)*100)
    
```
